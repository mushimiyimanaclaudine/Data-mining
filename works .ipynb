```python
import pandas as pd
df=pd.read_csv('financial_sample.csv')
df
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Print cleaned column names
print(df.columns)
```

```python

new_dataset = df[['segment','country','product', 'discount_band', 'units_sold','sale_price','sales','date','month_name', 'year']]
new_dataset.to_csv("new_dataset.csv", index=False)
new_dataset.head()
```

```python
df_renamed = new_dataset.rename(columns={
    "segment": "customer_name",
    "country": "city",
    "product": "product_name",
    "discount_band":"category",
    "units_sold":"quantity",
    "sale_price":"price",
    "sales":"total_sales",
    "month_name":"month"
    
})

# Display updated column names
print(df_renamed.columns)
```

```python
df_renamed
```

```python
df_renamed['price'] = df_renamed['price'].str.replace(r'\$', '', regex=True)
df_renamed['total_sales'] = df_renamed['total_sales'].str.replace(r'\$', '', regex=True)
df_renamed
```

```python
df_renamed["quantity"] = pd.to_numeric(df_renamed["quantity"], errors="coerce")
df_renamed["price"] = pd.to_numeric(df_renamed["price"], errors="coerce")
df_renamed
```

```python
df_renamed["total_sales"] = df_renamed["quantity"] * df_renamed["price"]
```

```python
df_renamed["date"] = pd.to_datetime(df_renamed["date"])
```

```python
df_renamed
```

```python
customer_data = df_renamed[['customer_name','city']]
customer_data.to_csv("customer_data.csv", index=False)
```

```python
date_data = df_renamed[['date','month','year']]
date_data.to_csv("date_data.csv", index=False)
```

```python
product_data = df_renamed[['product_name','category']]
product_data.to_csv("product_data.csv", index=False)
```

```python
sales_data=df_renamed[['quantity','price','total_sales']]
sales_data.to_csv("sales_data.csv",index=False)
```

```python
import pandas as pd
import mysql.connector
import csv

# Connect to MySQL database
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="sales"
)

cursor = mydb.cursor()

# Confirm database connection
cursor.execute("SELECT DATABASE();")
db_name = cursor.fetchone()
print(f"Connected to database: {db_name[0]}")

# Open CSV file and insert data
filename = "sales_data.csv"
with open(filename, 'r', newline='', encoding='utf-8') as csvfile:
    csvread = csv.reader(csvfile)
    next(csvread)  # Skip header

    for row in csvread:
        if len(row) < 6:  # Ensure row has all required columns
            print(f"Skipping row due to missing values: {row}")
            continue

        product_id, customer_id, date_id, quantity, price, total_sales = row[:6]  # Extract first 6 columns safely

        # Check if product_id, customer_id, and date_id exist in the referenced tables
        cursor.execute("SELECT product_id FROM product_data WHERE product_id=%s", (product_id,))
        product_exists = cursor.fetchone()

        cursor.execute("SELECT customer_id FROM customer_data WHERE customer_id=%s", (customer_id,))
        customer_exists = cursor.fetchone()

        cursor.execute("SELECT date_id FROM date_data WHERE date_id=%s", (date_id,))
        date_exists = cursor.fetchone()
        

        # Insert data only if all foreign keys exist
        if product_exists and customer_exists and date_exists:
            sql = """INSERT INTO sales_data (product_id, customer_id, date_id, quantity, price, total_sales)
                     VALUES (%s, %s, %s, %s, %s, %s)"""
            val = (product_id, customer_id, date_id, quantity, price, total_sales)
            cursor.execute(sql, val)
        else:
            print(f"Skipping row due to missing foreign key references: {row}")

# Commit changes and close connections
mydb.commit()
cursor.close()
mydb.close()
```
